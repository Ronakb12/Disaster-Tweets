```python
from transformers import AutoModelForCausalLM, AutoTokenizer, set_seed
```

```python
import torch
torch.set_default_tensor_type(torch.cuda.FloatTensor)
```

```python
model = AutoModelForCausalLM.from_pretrained('bigscience/bloom-1b7', use_cache=True)
tokenizer = AutoTokenizer.from_pretrained("bigscience/bloom-1b7")
```

```python
set_seed(2001)
```

```python
model_in_use = model
prompt='*Passage 1*\nArtificial intelligence is intelligence demonstrated by machines, as opposed to the natural intelligence displayed by animals and humans.\n*Passage 2*\nArtificial Intelligence is the field of developing computers and robots that are capable of behaving in ways that both mimic and go beyond human capabilities.\n\nQuestion : Are the two Passage same or different?\nAnswer :'
```

```python
input_ids = tokenizer(prompt, return_tensors="pt").to(0)
```

```python
sample = model_in_use.generate(**input_ids, max_length=200,  top_k=1, temperature=0.9 )
```

```python
print(tokenizer.decode(sample[0], truncate_before_pattern=[r"\n\n^#", "^'''", "\n\n\n"]))
```
